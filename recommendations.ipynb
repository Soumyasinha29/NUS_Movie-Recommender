```python
from ast import literal_eval
# Returns the list top l elements or entire list; whichever is more.
def get_list(x, l=5):
    if isinstance(x, list):
        names = [i['name'] for i in x]
        #Check if more than l elements exist. If yes, return only first three. If no, return entire list.
        if len(names) > l:
            names = names[:l]
        return names

    #Return empty list in case of missing/malformed data
    return []

movies_metadata_df['genres'] = movies_metadata_df['genres'].apply(literal_eval)
movies_metadata_df['genres'] = movies_metadata_df['genres'].apply(get_list)
```

```python
def check_missing(data_df):
    total = data_df.isnull().sum().sort_values(ascending = False)
    percent = (data_df.isnull().sum()/data_df.isnull().count()*100).sort_values(ascending = False)
    return pd.concat([total, percent], axis=1, keys=['Total', 'Percent']).transpose()

check_missing(ratings_df)
```

RECOMMENDATIONS BASED ON MOVIE TITLE:

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english',max_features=10000)
tokens = m_df[['title']]
tokens['title'] = tokens['title'].fillna('')
tfidf_matrix = tfidf.fit_transform(tokens['title'])
print(tfidf_matrix.shape)
from sklearn.metrics.pairwise import linear_kernel
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
print(cosine_sim.shape)
indices = pd.Series(tokens.index, index=tokens['title']).drop_duplicates()
def get_recommendations(title, cosine_sim=cosine_sim):
    # index of the movie that matches the title
    idx = indices[title]

    # similarity scores of all movies with that movie
    sim_scores = list(enumerate(cosine_sim[idx]))

    # sort the movies based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # scores of the 10 most similar movies
    sim_scores = sim_scores[1:11]

    # movie indices
    movie_indices = [i[0] for i in sim_scores]

    # top 10 most similar movies
    return tokens['title'].iloc[movie_indices]
```

```python
def get_imdb_score(df, indices):
    # select the data from similarity indices
    tmp = df[df.id.isin(indices)]
    # sort the data by IMDB score
    tmp = tmp.sort_values(by='IMDB', ascending=False)
    # return title and IMDB score
    return tmp[['title','IMDB']].head(10)
```

```python
def get_10_recommendations_simpol(title, cosine_sim=cosine_sim):
    # index of the movie that matches the title
    idx = indices[title]

    # similarity scores of all movies with that movie
    sim_scores = list(enumerate(cosine_sim[idx]))

    # sort the movies based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # scores of the 20 most similar movies
    sim_scores = sim_scores[1:21]
    
    # movie indices
    movie_indices = [i[0] for i in sim_scores]

    # get popularity scores
    pop_scores = get_imdb_score(m_df, movie_indices)
    
    return list(pop_scores['title'])
```

```python
from tkinter import *
import sys


def initGUI():
    WINDOW_WIDTH = 900
    WINDOW_HEIGHT = 500
    FONT = ("Myriad Pro", 13)
    FONT_BOLD = ("Myriad Pro", 13, 'bold')
    LARGE_FONT = ("Myriad Pro", 25)

    BG_COLOR = '#cfcfcf'
    WINDOW_TITLE = 'Movie Ratings'
    HEADING = 'Movie Recommendation System'

    GENRE = {1: 'Action', 2: 'Sci-Fi', 3: 'Comedy', 4: 'Adventure',
             5: 'Drama', 6: 'Romance', 7: 'Animated', 8: 'Fantsy', 9: 'Child', 10: 'Mystery' , 11: 'Thriller' , 12: 'Crime'}

    MOVIE_LIST = {
        'Deadpool (2016)': [1, 2], 'The Fault in Our Stars (2014)': [5, 3],
        'Toy Story (1995)': [4, 8], 'Shutter Island (2010)': [10, 11], 'Sherlock Holmes: The Woman in Green (1945)': [10],
        'Insidious: Chapter 2 (2013)': [11, 10], '21 Jump Street (2012)': [1, 3], 'Now You See Me 2 (2016)': [10, 12],
        'Dunkirk (2017)': [5, 10], 'The Fate of the Furious (2017)': [11, 12]}

    ratings = {}

    root = Tk()
    root.resizable(width=FALSE, height=FALSE)
    root.geometry('%sx%s' % (WINDOW_WIDTH, WINDOW_HEIGHT))
    root.config(bg=BG_COLOR)
    root.title(WINDOW_TITLE)

    for movie in MOVIE_LIST:
        ratings[movie] = IntVar()

    # Heading for Interface
    heading = Label(root, text=HEADING, font=LARGE_FONT,
                    fg='black', bg=BG_COLOR).pack(fill=BOTH, pady=20)

    # Separate Frame for all Movies and radio-buttons
    movie_frame = Frame(root, bg=BG_COLOR, width=.9 *
                        WINDOW_WIDTH, height=500, borderwidth=2, relief=GROOVE)
    movie_frame.pack(pady=20)

    # Movie Frame Heading
    frame_heading = Label(
        movie_frame, text='Movie', font=FONT_BOLD, bg=BG_COLOR)
    frame_heading.grid(row=0, column=0, pady=10)
    Label(movie_frame, text='Genre', font=FONT_BOLD,
          bg=BG_COLOR).grid(row=0, column=1, padx=50)
    for rating in range(1, 6):
        Label(movie_frame, text=str(rating), font=FONT_BOLD, bg=BG_COLOR).grid(
            row=0, column=rating + 1)
    Label(movie_frame, text='NA', font=FONT_BOLD,
          bg=BG_COLOR).grid(row=0, column=7)

    # List of Movie along with radio-buttons
    row = 1
    for movie in MOVIE_LIST:
        # Movie Label
        Label(movie_frame, text=movie, font=FONT,
              fg='black', bg=BG_COLOR).grid(row=row, sticky=W, padx=10, pady=(0, 5))

        # Genre Label
        genre = ''
        for val in MOVIE_LIST[movie]:
            genre += GENRE[val] if genre == '' else ', ' + GENRE[val]
        Label(movie_frame, text=genre, font=FONT, fg='black',
              bg=BG_COLOR).grid(row=row, column=1, sticky=W)

        # Radio Buttons for the Movie
        for rating in range(1, 6):
            Radiobutton(movie_frame, text='', variable=ratings[movie], value=rating, bg=BG_COLOR).grid(
                row=row, column=rating + 1, padx=((10, 2) if rating == 1 else 2))
        Radiobutton(movie_frame, text='', variable=ratings[movie], value=0, bg=BG_COLOR).grid(
            row=row, column=7, padx=(2, 10))

        row += 1

    Submit_button = Button(
        root, text="Submit", font=FONT, command=root.quit)
    Submit_button.pack(side=RIGHT, padx=(20, 40))
    Cancel_button = Button(
        root, text="Cancel", font=FONT, command=root.quit)
    Cancel_button.pack(side=RIGHT)
    # Cancel and Submit have the same function here

    root.mainloop()
    root.destroy()

    # Extracting values from IntVar() objects
    for movie in MOVIE_LIST:
        ratings[movie] = ratings[movie].get()
    return ratings


if __name__ == '__main__':
    ratings = initGUI()
    print(ratings)
```
